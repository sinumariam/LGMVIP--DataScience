#### #Task 4 - IMAGE TO PENCIL SKETCH WITH PYTHON (BEGINNER LEVEL TASK)
#### Problem Statement:
* Develop a Python Programme to get the Pencil Sketch of a RGB Image

```python
# Importing necessary libraries 
# OpenCV is a library for image processing,computer vision, machine learning
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')
```

```python
# Input Image is read
Input_image = cv2.imread("C:/Users/sinun/OneDrive/Documents/letsgrowmore/pencilsketch/plant.jpeg")
Input_image= cv2.cvtColor(Input_image,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(Input_image)
print("  INPUT RGB IMAGE   ")
plt.show()
```

#### * Input Image coverted to Grey Scale Image

```python
# Coverting Input Image to Grey Scale Image
grey_Scale_image = cv2.cvtColor(Input_image, cv2.COLOR_BGR2GRAY)
print("      GREY SCALE IMAGE   ")
display(Image.fromarray(grey_Scale_image))
```

#### * Grey Scale Image coverted to Inverted Image

```python
# Coverting Grey Scale Image to Inverted Image
Inverted_image = cv2.bitwise_not(grey_Scale_image)
print("      INVERTED IMAGE   ")
display(Image.fromarray(Inverted_image))
```

#### * Inverted Image converted into Blurred image


```python
# Converting Negative Image into Blurred image
Blurr_image = cv2.GaussianBlur(Inverted_image, (21, 21), 0)
print("      BLURRED IMAGE   ")
display(Image.fromarray(Blurr_image))
```

#### Blurred Image converted to Inverted Blurred Image


```python
Inverted_Blurr = cv2.bitwise_not(Blurr_image)
print("      INVERTED BLURRED IMAGE   ")
display(Image.fromarray(Inverted_Blurr))
```

#### Dividing Grey Scale image with Inverted Blurry Image to Get Output Pencil Sketch

```python
# Dividing Grey Scale Image with Inverted Blurred Image to get Output Pencil Sketch Image
Sketch_Output = cv2.divide(grey_Scale_image,Inverted_Blurr, scale=250.0)
print("          PENCIL SKETCH OUTPUT IMAGE        ")
display(Image.fromarray(Sketch_Output))
```
